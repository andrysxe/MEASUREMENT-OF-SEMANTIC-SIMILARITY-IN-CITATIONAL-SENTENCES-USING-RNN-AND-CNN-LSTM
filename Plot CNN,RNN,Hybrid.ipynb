```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Plot Accuracy Optimizer Adam di metode CNN, RNN, dan Hybrid CNN-LSTM

```python
log_data_cnn = pd.read_csv('cnnadam.log', sep=',', engine='python') 
log_data_rnn = pd.read_csv('6lblsg300adamw2v.log', sep=',', engine='python')
log_data_hybrid = pd.read_csv('hybridadam.log', sep=',', engine='python')
plt.rcParams.update({'font.size': 19})
fig = plt.figure() 
fig.set_size_inches(15,5)
# Plot the data
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
#plt.plot(primes,history.history['acc'], label='acc', color=new_colors[0],linewidth=2)
#plt.plot(log_data_cnn['acc'], label='acc', color='r',linewidth=2.2, marker='')
#plt.plot(log_data_cnn['val_acc'], label='val_acc', color='g',linewidth=2.2)
plt.plot(log_data_rnn['accuracy'], label='acc', color='maroon',linewidth=2.2)
plt.plot(log_data_rnn['val_accuracy'], label='val_acc', color='lime',linewidth=2.2)
plt.plot(log_data_hybrid['accuracy'], label='acc', color='gold',linewidth=2.2)
plt.plot(log_data_hybrid['val_accuracy'], label='val_acc', color='navy',linewidth=2.2)
plt.legend(['Adam RNN','Val Adam RNN','Adam Hybrid','Val Adam Hybrid'],loc='center left', bbox_to_anchor=(1, 0.5))
# Show the plot
plt.show()

#plot loss
fig = plt.figure() 
fig.set_size_inches(15,5)
# Plot the data
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
#plt.plot(primes,history.history['acc'], label='acc', color=new_colors[0],linewidth=2)
#plt.plot(log_data_cnn['loss'], label='loss', color='r',linewidth=2.2)
#plt.plot(log_data_cnn['val_loss'], label='val_loss', color='g',linewidth=2.2)
plt.plot(log_data_rnn['loss'], label='loss', color='maroon',linewidth=2.2)
plt.plot(log_data_rnn['val_loss'], label='val_loss', color='lime',linewidth=2.2)
plt.plot(log_data_hybrid['loss'], label='loss', color='gold',linewidth=2.2)
plt.plot(log_data_hybrid['val_loss'], label='val_loss', color='navy',linewidth=2.2)
plt.legend(['Adam RNN','Val Adam RNN','Adam Hybrid','Val Adam Hybrid'],loc='center left', bbox_to_anchor=(1, 0.5))
# Show the plot
plt.show()

#plot precision
fig = plt.figure() 
fig.set_size_inches(15,5)
# Plot the data
plt.title('Model Precision')
plt.ylabel('Precision')
plt.xlabel('Epoch')
#plt.plot(primes,history.history['acc'], label='acc', color=new_colors[0],linewidth=2)
#plt.plot(log_data_cnn['prec'], label='prec', color='r',linewidth=2.2)
#plt.plot(log_data_cnn['val_prec'], label='val_prec', color='g',linewidth=2.2)
plt.plot(log_data_rnn['prec'], label='prec', color='maroon',linewidth=2.2)
plt.plot(log_data_rnn['val_prec'], label='val_prec', color='lime',linewidth=2.2)
plt.plot(log_data_hybrid['prec'], label='prec', color='gold',linewidth=2.2)
plt.plot(log_data_hybrid['val_prec'], label='val_prec', color='navy',linewidth=2.2)
plt.legend(['Adam RNN','Val Adam RNN','Adam Hybrid','Val Adam Hybrid'],loc='center left', bbox_to_anchor=(1, 0.5))
# Show the plot
plt.show()

#plot recall
fig = plt.figure() 
fig.set_size_inches(15,5)
# Plot the data
plt.title('Model Recall')
plt.ylabel('Recall')
plt.xlabel('Epoch')
#plt.plot(primes,history.history['acc'], label='acc', color=new_colors[0],linewidth=2)
#plt.plot(log_data_cnn['rec'], label='rec', color='r',linewidth=2.2)
#plt.plot(log_data_cnn['val_rec'], label='val_rec', color='g',linewidth=2.2)
plt.plot(log_data_rnn['rec'], label='rec', color='maroon',linewidth=2.2)
plt.plot(log_data_rnn['val_rec'], label='val_rec', color='lime',linewidth=2.2)
plt.plot(log_data_hybrid['rec'], label='rec', color='gold',linewidth=2.2)
plt.plot(log_data_hybrid['val_rec'], label='val_rec', color='navy',linewidth=2.2)
plt.legend(['Adam RNN','Val Adam RNN','Adam Hybrid','Val Adam Hybrid'],loc='center left', bbox_to_anchor=(1, 0.5))
# Show the plot
plt.show()
```
